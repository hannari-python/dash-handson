```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
df = pd.DataFrame(iris.data)
df1 = pd.DataFrame(iris.target)
```

```python
df
```

```python
df1
```

```python
df = pd.concat([df, df1], axis=1)
```

```python
df
```

```python
colname = iris.feature_names
```

```python
colname
```

```python
colname.append('speces_num')
```

```python
colname
```

```python
df.columns = colname
```

```python
df
```

```python
iris.target_names
```

```python
def num_to_name(x):
    if x == 0:
        return iris.target_names[0]
    elif x == 1:
        return iris.target_names[1]
    else:
        return iris.target_names[2]
    
df['speces'] = df.apply(lambda row: num_to_name(row['speces_num']), axis=1)
```
